```python
import os 
def delete_file(file_path):

    if os.path.exists(file_path):
        if os.access(file_path, os.R_OK) and os.access(file_path, os.W_OK):
            os.remove(file_path)
            print("File has been deleted")
        else:
            print("No access to delete")

    else:
        print("File does not exist")

file1 = "A.txt"
delete_file(file1)
```
