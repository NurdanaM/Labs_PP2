```python
def countLines(filename):
    try:
        with open(filename, 'r') as file:
            lines = file.readlines()
            return len(lines)
        
    except FileNotFoundError:
        print("The file does nor exist")
        return 0
    
file_name = "/Users/nurdanam/Desktop/all_labs_pp2/Lab6/dirandfiles/test.txt"
lines_num = countLines(file_name)
print("Number of lines: ", lines_num)
```
